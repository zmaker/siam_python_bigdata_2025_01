# Grafici mathplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
numeri = np.random.randint(10,100,30)
print(numeri)
```

```python
_, ax = plt.subplots()
x = np.arange(0, len(numeri))
ax.plot(x, numeri)
plt.show()
```

# Grafico a torta

```python
valori =[15, 30, 45, 10, 25] 
nome = ["a", "b", "c", "d", "e"]
colori = ["#ff9999", "#66b3ff", '#3258a8', '#22a8a8', '#4832a8']


plt.pie(valori, labels=nome, colors=colori, autopct="%1.1f%%", startangle=90)
plt.show()
```

# scatter

```python
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
