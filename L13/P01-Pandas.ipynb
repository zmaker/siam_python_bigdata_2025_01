# Join

```python
data = {
    'codice': range(1001, 1006),
    'nome': ['mela', 'pera','kiwi','banana','fragola']
}
df_prodotti = pd.DataFrame(data)
df_prodotti
```

```python
data = {
    'codice': range(1001, 1011),
    'qta': [10,20,30,40,50,60,70,80,90,78]
}
df_magazzino = pd.DataFrame(data)
df_magazzino
```

```python
pd.merge(df_prodotti, df_magazzino, on='codice', how='left')
```

```python
pd.merge(df_prodotti, df_magazzino, on='codice', how='right')
```

# SQL e sqlite con SQLAlchemy 

```python
from sqlalchemy import create_engine
eng = create_engine('sqlite:///database.db')
```

```python
c = eng.connect()
df1 = pd.read_sql_table("coffee", c)
c.close()
df1
```

```python
eng = create_engine('sqlite:///prodotti.db')
df_prodotti.to_sql('prodotti', eng, index=False)
```
