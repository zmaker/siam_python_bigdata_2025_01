La funzione get_dummies in Pandas viene utilizzata per la codifica one-hot encoding, trasformando le colonne categoriali di un DataFrame in colonne binarie (0 e 1). Questo è utile per convertire variabili categoriali in un formato numerico, spesso richiesto da modelli di machine learning.

Parametri principali:
- data: Il DataFrame o la Serie da convertire.
- columns: Specifica le colonne categoriali da convertire (se non specificato, converte tutte quelle categoriali).
- prefix: Aggiunge un prefisso ai nomi delle colonne risultanti.
- drop_first: Se impostato a True, elimina la prima colonna dummy per evitare la dummy variable trap (collinearità tra le colonne).
- dtype: Specifica il tipo di dato delle colonne risultanti (default int).

```python
#Dummies / One-Hot Encoding
# NEAR BAY   INLAND   NEAR OCEAN
#    1          0         0
#    0          0         1

dummies = pd.get_dummies(datos_na["ocean_proximity"], dtype=int)
dummies.head()
```

```python
datos_na = datos_na.join(dummies)
```

```python
datos_na.head()
```

```python
datos_na = datos_na.drop(["ocean_proximity"], axis=1)
```
